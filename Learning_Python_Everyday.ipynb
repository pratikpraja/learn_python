# 1. Bubble sort Algorithm: 

bubblesort.png
Image credit: https://medium.com/karuna-sehgal/an-introduction-to-bubble-sort-d85273acfcd8

```python
# Bubble sort 

def bubblesort(list1):
    n=len(list1)
    
    for i in range(n):
        for j in range(0,n-i-1):
            if list1[j]>list1[j+1]:
                temp = list1[j]
                list1[j]=list1[j+1]
                list1[j+1]=temp

list1=[]         # creating an empty
n=int(input("Enter number of elements: "))

for k in range(0,n):
    element = int(input())
    
    list1.append(element)
    
print(list1)

bubblesort(list1)

print ("Sorted array is:", list1)
```

# 2. Insertion sort Algorithm

insertion-sort.png
Image credit: https://www.w3resource.com/csharp-exercises/searching-and-sorting-algorithm/searching-and-sorting-algorithm-exercise-6.php

```python
# Insertion sort

# hP = Hole Position
# vI = Value to insert
def insertionSort(list1):
    for i in range(1,len(list1)):
        vI = list1[i]
        hP = i
        
        while hP>0 and list1[hP-1]>vI:
            list1[hP]=list1[hP-1]
            hP=hP-1
        list1[hP]=vI

list1=[]         # creating an empty
n=int(input("Enter number of elements: "))

for k in range(0,n):
    element = int(input())
    
    list1.append(element)
    
print(list1)

insertionSort(list1)

print ("Sorted array is:", list1)
```

# 3. Selection sort Algorithm:

selection-sort.png
Image credit: https://www.w3resource.com/php-exercises/searching-and-sorting-algorithm/searching-and-sorting-algorithm-exercise-4.php

```python
# Selection sort

def selectionSort(list1):
    n=len(list1)
    for i in range(0,n-1):
        minN = i
        for j in range(i+1,n):
            if list1[j]<list1[minN]:
                minN = j
        if minN != i:
            list1[i],list1[minN]=list1[minN],list1[i]

list1=[]         # creating an empty
n=int(input("Enter number of elements: "))

for k in range(0,n):
    element = int(input())
    
    list1.append(element)
    
print(list1)

selectionSort(list1)

print ("Sorted array is:", list1)
```
